**This file performs operations on our original dataset to filter out specific genres, clean the data, and export it to a new dataset csv file. Also provides some data visualizations describing the dataset.**

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

## Filtering Genres

```python
data = pd.read_csv('one_genre_with_encoding.csv', usecols = lambda x: x.startswith('Unnamed') == False)
```

```python
data.replace({'\n': ' '}, regex = True, inplace = True)
```

```python
data['Word Count'] = data.Lyric.str.split().str.len()
```

```python
data
```

```python
sns.kdeplot(data['Word Count'], fill = True)
```

## Filter by Word Count
Only use songs with 50 + words and songs with < 900 words

```python
data = data[(data['Word Count'] >= 50) & (data['Word Count'] <= 900)]
```

```python
data
```

```python
ax = sns.countplot(data=data, x= "Genres")
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
```

```python
data[['Genres', 'Artist']].groupby(by='Genres').count()
```

```python
from sklearn.utils import resample
```

```python
balanced_df = pd.DataFrame(columns= data.columns)
genres = data['Genres'].unique()
smallest_sample_size = 5202

# downsize all classes to match the minimum
for genre in genres:
    resampled_data = resample(data[data['Genres'] == genre],
                              replace= False,
                              n_samples = smallest_sample_size,
                              random_state = 111)
    balanced_df = pd.concat([balanced_df, resampled_data])
```

```python
balanced_df = pd.DataFrame(columns= data.columns)
genres = data['Genres'].unique()
largest = 24793

# downsize all classes to match the minimum
for genre in genres:
    resampled_data = resample(data[data['Genres'] == genre],
                              replace= True,
                              n_samples = largest,
                              random_state = 111)
    balanced_df = pd.concat([balanced_df, resampled_data])
```

```python
balanced_df.to_csv("upsize_balanced_dataset.csv", index=False)
```

```python
lyrics_genre = pd.read_csv("../datasets/genre_lyrics.csv")
```

```python
lyrics_genre = lyrics_genre[['ALink', 'SName', 'SLink', 'Lyric', 'language', 'Artist', 'Genres', 'Songs', 'Popularity']]
```

```python
lyrics_genre
```

```python
top_10 = {"Country",
"Heavy Metal",
"Hip Hop",
"Indie",
"Pop",
"Pop/Rock",
"R&B",
"Rap",
"Rock",
"Rock Alternativo"}
```

```python
lyrics_genre['Genres'] = lyrics_genre['Genres'].astype(str)
```

```python
def checker(x):
    if x.split(";")[0] in top_10:
        return True
    else:
        return False
    
only_top_10 = lyrics_genre[lyrics_genre['Genres'].apply(checker)]
```

```python
only_top_10.to_csv("unbalanced_all_genres_top_10.csv", index = False)
```

```python
only_top_10['Top Genre'] = only_top_10['Genres'].apply(lambda x: x.split(';')[0])
```

```python
only_top_10
```

```python
balanced_df = pd.DataFrame(columns= data.columns)
genres = data['Genres'].unique()
smallest_sample_size = 5202

# downsize all classes to match the minimum
for genre in genres:
    resampled_data = resample(only_top_10[only_top_10['Top Genre'] == genre],
                              replace= False,
                              n_samples = smallest_sample_size,
                              random_state = 111)
    balanced_df = pd.concat([balanced_df, resampled_data])
```

```python
balanced_df[['ALink', 'SName', 'SLink', 'Lyric', 'language', 'Artist', 'Genres',
       'Songs', 'Popularity', 'Encoding', 'Word Count']].to_csv("balanced_all_genres_top_10.csv", index = False)
```
